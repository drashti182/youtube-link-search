```python
# First, install the library if you haven't
# pip install youtube-search-python

from youtubesearchpython import VideosSearch

def get_youtube_links(query, limit=5):
    """
    Returns a list of YouTube video links for a search query.
    
    :param query: Search string
    :param limit: Number of results to return
    :return: List of YouTube URLs
    """
    video_search = VideosSearch(query, limit=limit)
    results = video_search.result().get('result', [])
    
    links = []
    for video in results:
        links.append(video['link'])
    
    return links

# Example usage
search_query = "Python programming tutorial"
links = get_youtube_links(search_query)
print(links)
```
